```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage

from models.generations import ActorAgentGeneration

mini_llm = ChatOpenAI(temperature=0.7, model="gpt-4.1-mini").with_structured_output(ActorAgentGeneration)
not_mini_llm = ChatOpenAI(temperature=0.7, model="gpt-4.1").with_structured_output(ActorAgentGeneration)
```

```python
import asyncio


messages = [
    SystemMessage(system_prompt),
    HumanMessage("generate a succinct summary of any outcomes and what Jade's next action will be")
]

mini_futures = [mini_llm.ainvoke(messages) for i in range(3)]
not_mini_futures = [not_mini_llm.ainvoke(messages) for i in range(3)]
mini_responses = await asyncio.gather(*mini_futures)
not_mini_responses = await asyncio.gather(*not_mini_futures)

print("GPT 4.1 MINI")
for response in mini_responses:
    print(response)
    print("##############################################")

print("##############################################")
print("##############################################")
print("##############################################")
print("GPT 4.1 (NOT MINI)")
for response in not_mini_responses:
    print(response)
    print("##############################################")
```
