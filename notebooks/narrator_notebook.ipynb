```python
# INIT LLM CLIENT

from langchain_openai import ChatOpenAI

llm = ChatOpenAI(temperature=0.9, model="gpt-4.1")
```

```python
# LOAD GAME WORLD, GENERATE NARRATIVE, SAVE NARRATIVE

import asyncio
import dill
from langchain_core.messages import HumanMessage, SystemMessage

from prompts.narrator_agent_prompts import build_narrator_agent_prompt

with open("sandbox/processed_game_world_1.pkl", "rb") as f:
    game_world = dill.load(f)


messages = [
    SystemMessage(build_narrator_agent_prompt(game_world, 0)),
    HumanMessage("Write a compelling narratve based on the above instructions")
]

futures = [llm.ainvoke(messages) for i in range(1)]
responses = await asyncio.gather(*futures)

with open("sandbox/processed_game_world_narrative_1.pkl", "wb") as f:
    dill.dump(responses[0].content, f)
```

```python
# LOAD AND PRINT A NARRATIVE

import dill

with open("sandbox/processed_game_world_narrative_1.pkl", "rb") as f:
    narrative = dill.load(f)

print(narrative)
```
